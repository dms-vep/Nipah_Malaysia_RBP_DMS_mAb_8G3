```python
import pandas as pd
```

```python
df = pd.read_csv('../results/antibody_escape/averages/8G3_mut_effect.csv')
func_effects_df = pd.read_csv('../results/func_effects/averages/CHO_bEFNB3_func_effects.csv')
display(df, func_effects_df)
```

```python
merged_df = pd.merge(df[['site','wildtype','mutant','mutation','escape_mean','escape_std','times_seen']], func_effects_df, on=['site','wildtype','mutant'], suffixes=('_antibody','_func_effects'))
display(merged_df)
```

```python
filtered_df = merged_df[
    (merged_df['times_seen_antibody'] >= 2)
    & (merged_df['effect'] >= -2)
    & (merged_df['effect_std'] <= 3)
]
filtered_df = filtered_df.drop(columns=['n_selections', 'times_seen_func_effects'])
filtered_df = filtered_df.rename(columns={'effect':'func_effect','effect_std':'func_effect_std'})
display(filtered_df)
```

```python
filtered_df.to_csv('../results/filtered_data/8G3_mut_effect_filtered.csv', index=False)
```
